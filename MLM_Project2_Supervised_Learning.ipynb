```python
# Handling Missing Data
def missing_data_summary(df):
    """
    Creates a DataFrame summarizing missing data statistics.

    Args:
        df: The input DataFrame.

    Returns:
        A DataFrame with missing data count and percentage for each variable and row.
    """
    # Column-wise summary
    missing_count_col = df.isnull().sum()
    missing_percentage_col = (missing_count_col / len(df)) * 100
    missing_summary_col = pd.DataFrame({
        'Missing Count': missing_count_col,
        'Missing Percentage': missing_percentage_col
    })
    # Row-wise summary
    missing_count_row = df.isnull().sum(axis=1)
    missing_percentage_row = (missing_count_row / len(df.columns)) * 100
    missing_summary_row = pd.DataFrame({
        'Missing Count': missing_count_row,
        'Missing Percentage': missing_percentage_row
    }, index=df.index)  # Use original DataFrame index

    return missing_summary_col, missing_summary_row

# Calculate missing data summary for categorical subset
sr40_missing_summary_cat_col, sr40_missing_summary_cat_row = missing_data_summary(sr40_ds_cat)

# Calculate missing data summary for non-categorical subset
sr40_missing_summary_noncat_col, sr40_missing_summary_noncat_row = missing_data_summary(sr40_ds_noncat)

# Drop columns and rows with more than 50% missing data for categorical subset
sr40_ds_cat_treated = sr40_ds_cat.dropna(thresh=len(sr40_ds_cat) * 0.5, axis=1)
sr40_ds_cat_treated = sr40_ds_cat_treated.dropna(thresh=len(sr40_ds_cat_treated.columns) * 0.5, axis=0)

# Drop columns and rows with more than 50% missing data for non-categorical subset
sr40_ds_noncat_treated = sr40_ds_noncat.dropna(thresh=len(sr40_ds_noncat) * 0.5, axis=1)
sr40_ds_noncat_treated = sr40_ds_noncat_treated.dropna(thresh=len(sr40_ds_noncat_treated.columns) * 0.5, axis=0)

# Calculate total variables and rows dropped
sr40_dropped_variables_cat = len(sr40_ds_cat.columns) - len(sr40_ds_cat_treated.columns)
sr40_dropped_rows_cat = len(sr40_ds_cat) - len(sr40_ds_cat_treated)
sr40_dropped_variables_noncat = len(sr40_ds_noncat.columns) - len(sr40_ds_noncat_treated.columns)
sr40_dropped_rows_noncat = len(sr40_ds_noncat) - len(sr40_ds_noncat_treated)

sr40_total_dropped_variables = sr40_dropped_variables_cat + sr40_dropped_variables_noncat
sr40_total_dropped_rows = sr40_dropped_rows_cat + sr40_dropped_rows_noncat

# Print missing data summaries
print("Categorical Missing Data Summary (Columns):")
print(sr40_missing_summary_cat_col)
print("\nCategorical Missing Data Summary (Rows):")
print(sr40_missing_summary_cat_row)
print("\nNon-Categorical Missing Data Summary (Columns):")
print(sr40_missing_summary_noncat_col)
print("\nNon-Categorical Missing Data Summary (Rows):")
print(sr40_missing_summary_noncat_row)

# Print total dropped variables and rows
print(f"\nTotal variables dropped: {sr40_total_dropped_variables}")
print(f"Total rows dropped: {sr40_total_dropped_rows}")
```

```python
# Descriptive statistics for categorical variables
def sr40_categorical_stats(df, cat_columns):
    sr40_results = []

    for sr40_col in cat_columns:
        sr40_counts = df[sr40_col].value_counts()
        sr40_proportions = sr40_counts / len(df)
        sr40_mode = sr40_counts.idxmax()
        sr40_min = df[sr40_col].min()
        sr40_max = df[sr40_col].max()

        for category, count in sr40_counts.items():
            sr40_results.append({
                "Column": sr40_col,
                "Category": category,
                "Frequency": count,
                "Proportion": sr40_proportions[category],
                "Minimum": sr40_min,
                "Maximum": sr40_max,
                "Mode": sr40_mode
            })

    return pd.DataFrame(sr40_results)

# Function to calculate correlation for categorical columns
def sr40_categorical_correlation(df, cat_columns, method='spearman'):
    sr40_corr_results = pd.DataFrame(index=cat_columns, columns=cat_columns, dtype=float)

    for i in range(len(cat_columns)):
        for j in range(i, len(cat_columns)):
            sr40_col1, sr40_col2 = cat_columns[i], cat_columns[j]
            if i == j:
                # Perfect correlation for the same variable
                corr = 1.0
            else:
                if method == 'spearman':
                    corr, _ = spearmanr(df[sr40_col1].astype('category').cat.codes,
                                        df[sr40_col2].astype('category').cat.codes)
                elif method == 'kendall':
                    corr, _ = kendalltau(df[sr40_col1].astype('category').cat.codes,
                                         df[sr40_col2].astype('category').cat.codes)
            sr40_corr_results.loc[sr40_col1, sr40_col2] = corr
            sr40_corr_results.loc[sr40_col2, sr40_col1] = corr

    return sr40_corr_results

# Retrieve categorical columns from hb11_ds_cat_encoded
sr40_cat_columns = sr40_ds_cat_encoded.columns.tolist()

# Generate descriptive stats and correlations for these columns
sr40_categorical_stats_df = sr40_categorical_stats(sr40_ds_preprocessed, sr40_cat_columns)
sr40_spearman_corr_df = sr40_categorical_correlation(sr40_ds_preprocessed, sr40_cat_columns, method='spearman')
sr40_kendall_corr_df = sr40_categorical_correlation(sr40_ds_preprocessed, sr40_cat_columns, method='kendall')

# Replace NaN correlations with 1 for diagonal elements (if any remain)
sr40_spearman_corr_df.fillna(1, inplace=True)
sr40_kendall_corr_df.fillna(1, inplace=True)

# Display results
print("Descriptive Statistics:")
print(sr40_categorical_stats_df)  # Show all rows

print("\nSpearman Correlation:")
print(sr40_spearman_corr_df)

print("\nKendall Correlation:")
print(sr40_kendall_corr_df)
```

```python
# Descriptive Stats for Noncat Variables

# Function to calculate descriptive statistics for numerical variables
def sr40_numerical_stats(df, num_columns):
    sr40_results = []

    for sr40_col in num_columns:
        sr40_data = df[sr40_col].dropna()  # Exclude NaN values
        sr40_min = sr40_data.min()
        sr40_max = sr40_data.max()
        sr40_mean = sr40_data.mean()
        sr40_median = sr40_data.median()
        sr40_mode = sr40_data.mode().iloc[0] if not sr40_data.mode().empty else np.nan
        sr40_range = sr40_max - sr40_min
        sr40_std_dev = sr40_data.std()
        sr40_skewness = skew(sr40_data)
        sr40_kurt = kurtosis(sr40_data)

        # Coefficient of Variation (CV)
        sr40_cv = sr40_std_dev / sr40_mean if sr40_mean != 0 else np.nan

        # Confidence Interval (95%)
        sr40_conf_interval = norm.interval(0.95, loc=sr40_mean, scale=sr40_std_dev / np.sqrt(len(sr40_data)))

        # Add results
        sr40_results.append({
            "Column": sr40_col,
            "Minimum": sr40_min,
            "Maximum": sr40_max,
            "Mean": sr40_mean,
            "Median": sr40_median,
            "Mode": sr40_mode,
            "Range": sr40_range,
            "Standard Deviation": sr40_std_dev,
            "Skewness": sr40_skewness,
            "Kurtosis": sr40_kurt,
            "Coefficient of Variation": sr40_cv,
            "Confidence Interval (95%)": sr40_conf_interval
        })

    return pd.DataFrame(sr40_results)

# Function to calculate correlation for numerical variables
def sr40_numerical_correlation(df, num_columns, method='pearson'):
    sr40_corr_results = pd.DataFrame(index=num_columns, columns=num_columns, dtype=float)

    for i in range(len(num_columns)):
        for j in range(i, len(num_columns)):
            sr40_col1, sr40_col2 = num_columns[i], num_columns[j]
            if i == j:
                corr = 1.0  # Perfect correlation for the same variable
            else:
                if method == 'pearson':
                    corr, _ = pearsonr(df[sr40_col1].dropna(), df[sr40_col2].dropna())
                elif method == 'spearman':
                    corr, _ = spearmanr(df[sr40_col1].dropna(), df[sr40_col2].dropna())
            sr40_corr_results.loc[sr40_col1, sr40_col2] = corr
            sr40_corr_results.loc[sr40_col2, sr40_col1] = corr

    return sr40_corr_results

# Retrieve numerical columns from hb11_ds_noncat_scaled
sr40_num_columns = sr40_ds_noncat_scaled.columns.tolist()

# Generate statistics and correlations for these columns
sr40_numerical_stats_df = sr40_numerical_stats(sr40_ds_noncat_scaled, sr40_num_columns)
sr40_pearson_corr_df = sr40_numerical_correlation(sr40_ds_noncat_scaled, sr40_num_columns, method='pearson')
sr40_spearman_corr_df = sr40_numerical_correlation(sr40_ds_noncat_scaled, sr40_num_columns, method='spearman')

# Replace NaN correlations with 1 for diagonal elements (if any remain)
sr40_pearson_corr_df.fillna(1, inplace=True)
sr40_spearman_corr_df.fillna(1, inplace=True)

# Display results
print("Numerical Statistics:")
print(sr40_numerical_stats_df)

print("\nPearson Correlation:")
print(sr40_pearson_corr_df)

print("\nSpearman Correlation:")
print(sr40_spearman_corr_df)
```

## Inferential Statistics

```python
# Select the two categorical variables for the test
sr40_categorical_var1 = 'Import_Export'
sr40_categorical_var2 = 'Shipping_Method'

# Create a contingency table
sr40_contingency_table = pd.crosstab(sr40_ds_preprocessed[sr40_categorical_var1],
                                     sr40_ds_preprocessed[sr40_categorical_var2])

# Print the contingency table
print("Contingency Table:")
print(sr40_contingency_table)


# Perform the Chi-squared test
sr40_chi2_stat, sr40_p_value, sr40_dof, sr40_expected_freq = chi2_contingency(sr40_contingency_table)

# Print the results
print(f"Chi-squared statistic: {sr40_chi2_stat}")
print(f"P-value: {sr40_p_value}")
print(f"Degrees of freedom: {sr40_dof}")
print("Expected frequencies:")
print(sr40_expected_freq)

# Interpret the results
sr40_alpha = 0.05  # Significance level
if sr40_p_value < sr40_alpha:
    print("Reject the null hypothesis. There is a significant association between the variables.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between the variables.")
```

```python
# List of non-categorical variables
sr40_non_categorical_vars = ['Quantity', 'Value', 'Weight']

# Loop through each variable
for sr40_var in sr40_non_categorical_vars:
    print(f"\n--- Tests for {sr40_var} ---")

    # Shapiro-Wilk Test
    sr40_shapiro_stat, sr40_shapiro_p_value = stats.shapiro(sr40_ds_preprocessed[sr40_var])
    print(f"Shapiro-Wilk Test - Statistic: {sr40_shapiro_stat}, P-value: {sr40_shapiro_p_value}")
    if sr40_shapiro_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")

    # Kolmogorov-Smirnov Test
    sr40_ks_stat, sr40_ks_p_value = stats.kstest(sr40_ds_preprocessed[sr40_var], 'norm')
    print(f"Kolmogorov-Smirnov Test - Statistic: {sr40_ks_stat}, P-value: {sr40_ks_p_value}")
    if sr40_ks_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")

    # Anderson-Darling Test
    sr40_anderson_result = stats.anderson(sr40_ds_preprocessed[sr40_var])
    print("Anderson-Darling Test - Statistic:", sr40_anderson_result.statistic)
    # Check only for 5% significance level
    if sr40_anderson_result.statistic < sr40_anderson_result.critical_values[2]:  # Index 2 corresponds to 5%
        print(f"Data looks normal (fail to reject H0) at 5% significance level")
    else:
        print(f"Data does not look normal (reject H0) at 5% significance level")

    # Jarque-Bera Test
    sr40_jb_stat, sr40_jb_p_value = stats.jarque_bera(sr40_ds_preprocessed[sr40_var])
    print(f"Jarque-Bera Test - Statistic: {sr40_jb_stat}, P-value: {sr40_jb_p_value}")
    if sr40_jb_p_value < 0.05:
        print(f"Data does not look normal (reject H0) at 5% significance level")
    else:
        print(f"Data looks normal (fail to reject H0) at 5% significance level")
```
